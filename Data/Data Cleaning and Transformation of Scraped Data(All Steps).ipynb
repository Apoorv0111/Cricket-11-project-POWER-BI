```python
import json
import pandas as pd
```

##  1).RESULTS 

```python
with open('t20_wc_match_results.json') as f:
    data=json.load(f)
    
print(len(data))

x=data[0]['matchSummary']
df=pd.DataFrame(x)
df.head()

```

```python
df.shape
```

```python
df.rename({'scorecard':'match_id'}, axis=1,inplace=True)
df.head()
        
        
       
    
```

```python
df.to_csv('match_results.csv', index=False)
```

## 2). BATTING SUMMARY

```python
with open('t20_wc_batting_summary.json') as d:
    data1=json.load(d)
    all_records=[]
    for rec in data1:
        all_records.extend(rec['battingSummary'])
result=pd.DataFrame(all_records)
result.head(10)
```

```python
result.head(5)
```

```python
result.shape
```

```python
result['Out/Not_Out']=result['dismissal'].apply(lambda x: 'Out' if len(x)>0 else 'Not_Out')

result.head()
```

```python
result.drop(columns=['dismissal'], inplace=True)
```

```python
result.head()
```

```python
print(result['batsmanName'])
```

```python
result['batsmanName'] = result['batsmanName'].apply(lambda x: x.replace('(c)†',''))
result['batsmanName'] = result['batsmanName'].apply(lambda x: x.replace('†',''))
result['batsmanName'] = result['batsmanName'].apply(lambda x: x.replace('(c)',''))
```

```python
result.tail(20)
```

```python
match_dicts={}
for index, row in df.iterrows():
    key1=row['team1']+' Vs '+row['team2']
    key2=row['team2']+' Vs '+row['team1']

    match_dicts[key1]=row['match_id']
    match_dicts[key2]=row['match_id']
print(match_dicts)
    
```

```python
result['match_id']=result['match'].map(match_dicts)
```

```python
result.head(30)
```

```python
result.to_csv('batting_Summary.csv', index=False)
```

## 3). BOWLING SUMMARY

```python
with open('t20_wc_bowling_summary.json') as a:
    data2=json.load(a)
    all_records=[]
    for rec in data2:
        all_records.extend(rec['bowlingSummary'])
        
        
```

```python
df_bowling = pd.DataFrame(all_records)
df_bowling.head(30)
```

```python
df_bowling.shape
```

```python
df_bowling['match_id'] = df_bowling['match'].map(match_dicts)
df_bowling.head()
```

```python
df_bowling.to_csv('bowling_Summary.csv', index=False)
```
